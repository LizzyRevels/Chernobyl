```python
import os
import pandas as pd
```

```python
chernobyl = "CHERNAIR.csv"
```

```python
chernobyl_df = pd.read_csv(chernobyl, encoding="ISO-8859-1")
chernobyl_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
chernobyl_renamed_df = chernobyl_df.rename(columns={
    "PAYS":"Country", 
    "Code":"Country Code", 
    "Ville":"Locality Name",
    "X":"Latiitude", 
    "Y":"Longitude" , 
    "I 131 (Bq/m3)" : "Iodine-131 (I-131) Concentration (Bq/m3)" , 
    "Cs 134 (Bq/m3)" : "Caesium-134 (Cs-134) Concentration (Bq/m3)" , 
    "Cs 137 (Bq/m3)" : "Caesium-137 (radiocaesium, Cs-137) Concentration (Bq/m3)"})
chernobyl_renamed_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
chernobyl_renamed_df.to_csv("chernobyl.csv", index=False, header=True)
```
