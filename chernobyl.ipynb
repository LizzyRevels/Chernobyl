```python
import os
import pandas as pd
```

```python
chernobyl = "CHERNAIR.csv"
```

```python
# Read Kaggle csv
chernobyl_df = pd.read_csv(chernobyl, encoding="ISO-8859-1")
chernobyl_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
chernobyl_renamed_df = chernobyl_df.rename(columns={
    "PAYS":"Country", 
    "Code":"Country Code", 
    "Ville":"Locality Name",
    "X":"Latitude", 
    "Y":"Longitude" , 
    "I 131 (Bq/m3)" : "Iodine-131 (I-131) Concentration (Bq/m3)" , 
    "Cs 134 (Bq/m3)" : "Caesium-134 (Cs-134) Concentration (Bq/m3)" , 
    "Cs 137 (Bq/m3)" : "Caesium-137 (radiocaesium, Cs-137) Concentration (Bq/m3)"})
chernobyl_renamed_df.head()
```

```python
# Count incomplete rows
chernobyl_renamed_df.count()
```

```python
# Display an overview of the Country column
chernobyl_renamed_df['Country'].value_counts()
```

```python
# Replace country abbreviation with country name.
chernobyl_renamed_df['Country'] = chernobyl_renamed_df['Country'].replace(
    {'IT': 'Italy', 
     'F': 'France',
     'DE':'Denmark',
     'AU':'Australia',
     'SE':'Sweden',
     'CZ':'Czech Republic',
     'NL':'Netherlands',
     'UK':'United Kingdom',
     'FI':'Finland',
     'NO':'Norway',
     'BE':'Belgium',
     'GR':'Greece',
     'ES':'Spain',
     'CH':'Switzerland',
     'IR':'Iran',
     'HU':'Hungary'})
chernobyl_renamed_df
```

```python
# Drop null rows
chernobyl_drop_df = chernobyl_renamed_df.dropna(how='any')
chernobyl_drop_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
chernobyl_drop_df.to_csv("chernobyl.csv", index=False, header=True)
```
